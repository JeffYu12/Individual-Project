```python
import pandas as pd
import matplotlib.pyplot as plt

!wget -O Player_Stats.xlsx "https://github.com/JeffYu12/Group-Project-/raw/main/Player%20Stats.xlsx"

data = pd.read_excel("Player_Stats.xlsx", sheet_name=None)
data_kings = pd.read_excel("Player_Stats.xlsx", sheet_name='Shooting 22-23 vs Kings')
data_Hornets = pd.read_excel("Player_Stats.xlsx", sheet_name='Shooting 22-23 vs Hornets')
data_Cavaliers = pd.read_excel("Player_Stats.xlsx", sheet_name='Shooting 22-23 vs Cavaliers')
```

```python
data['Shooting 22-23 vs Kings']
```

```python
rows = [data_kings.iloc[i] for i in range(5)]

# Extract player names and statistics
player_names = [row['Name'] for row in rows]
statistics = {
    'FG%': [row['FG%'] for row in rows],
    '3P%': [row['3P%'] for row in rows],
    'FT%': [row['FT%'] for row in rows],
    '2P%': [row['2P%'] for row in rows],
}

# Set bar width and positions
bar_width = 0.15
x = range(len(player_names))

plt.figure(figsize=(12, 6))

# Create bars for each statistic
for i, (stat_name, stat_values) in enumerate(statistics.items()):
    bars = plt.bar([p + bar_width * i for p in x], stat_values, width=bar_width, label=stat_name)

    # Add values above each bar
    for bar in bars:
        yval = bar.get_height()
        plt.text(bar.get_x() + bar.get_width() / 2, yval, round(yval, 1),
                 ha='center', va='bottom', fontsize=10)

# Customize the plot
plt.xlabel("Player Name")
plt.ylabel("Percentage (%)")
plt.title("Top 5 Celtics Player's Shooting Statistics vs Kings")
plt.xticks([p + bar_width * (len(statistics) / 2 - 0.5) for p in x], player_names, rotation=45)
plt.legend(title='Statistics')
plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
